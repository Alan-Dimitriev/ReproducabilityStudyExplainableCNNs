```python
def token_processing(input_string):
    index=input_string.find("Service:")
    if index !=-1 :
        input_string = input_string[index:] 
        
    index=input_string.find("Dictated By")
    if index !=-1 :
        input_string = input_string[:index]
        
    index=input_string.find("Followup Instructions:")
    if index !=-1 :
        input_string = input_string[:index]
    
    input_string = input_string.replace("\n", " ")
    input_string = input_string.replace("*", " ")
    input_string = input_string.replace("?", " ")
    input_string = input_string.replace("[", " ")
    input_string = input_string.replace("]", " ")
    input_string = input_string.replace("First Name", " ")
    input_string = input_string.replace("Last Name", " ")
    input_string = input_string.replace("Doctor Name", " ")
    
    data = []
  
    # iterate through each sentence in the file
    for i in sent_tokenize(input_string):
        temp = []

        # tokenize the sentence into words
        for j in word_tokenize(i):
            if any(c.isalpha() for c in j.lower()) is False:
                continue
            else:
                data.append(j.lower())

        #data.append(temp)
        
    
    return data
```

```python
import tqdm
from tqdm import tqdm
```

```python
def find_valid_tokens(all_words, list_of_sets):
    valid_tokens = []
    for x in tqdm(all_words):
        count = 0
        for y in list_of_sets:
            if x in y:
                count += 1
            if count > 2:
                valid_tokens.append(x)
                break
    return valid_tokens
```

```python
valid_tokens = find_valid_tokens(all_words, list_of_sets)
```
